```python
df_domestic = df_factory[df_factory["판매 구분"] == "내수용"].copy()
```

```python
month_cols = [f"{i}월" for i in range(1, 13)]
df_melted = df_domestic.melt(
    id_vars=["국가", "차량 모델", "연도"],
    value_vars=month_cols,
    var_name="월",
    value_name="판매량"
)
```

```python
import numpy as np
df_melted["판매량"] = df_melted["판매량"].replace("-", np.nan).astype(int)
df_melted.dropna(subset=["판매량"], inplace=True)
df_melted.head(10)
```

```python
# 기아 불필요한 공장명(국가) 제거
df_kia.drop(columns=["공장명(국가)"], inplace=True, errors="ignore")
```

```python
df_kia
```

```python
month_cols = [f"{i}월" for i in range(1, 13)]
df_kia_melted = df_kia.melt(
    id_vars=["국가명", "차종", "연도"],
    value_vars=month_cols,
    var_name="월",
    value_name="판매량"
)
```

```python
df_kia_melted["판매량"] = (
    df_kia_melted["판매량"]
    .replace("-", np.nan)
    .str.replace(",", "", regex=False)
    .astype(float)
)

# 2. NaN 제거
df_kia_melted["판매량"] = df_kia_melted["판매량"].fillna("0").astype(float).astype(int)

# 3. float → int 변환 (결측치 없으니 OK)
df_kia_melted["판매량"] = df_kia_melted["판매량"].astype(int)
```

```python
df_eu=df_euus[df_euus["국가명"] == "EU"].copy()
# 2. 월별 컬럼 melt
month_cols = [f"{i}월" for i in range(1, 13)]
df_eu_melted = df_eu.melt(
    id_vars=["차량 모델", "연도", "국가명"],
    value_vars=month_cols,
    var_name="월",
    value_name="판매량"
)

# 3. 시장구분 컬럼 추가
df_eu_melted["국가"] = "서유럽"

# 4. 컬럼 정리 (현대 내수용과 동일하게)
df_eu_melted_final = df_eu_melted[["국가", "차량 모델", "연도", "월", "판매량"]]

# 결과 미리보기
df_eu_melted_final.head()
```

```python
df_melted_combined = pd.concat([df_melted, df_eu_melted_final], ignore_index=True)
```
