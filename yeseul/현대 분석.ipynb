```python
df_all=df_all.groupby(['국가', '차량 모델', '연도', '월'], as_index=False)['판매량'].sum()
```

```python
df_all
```

```python
# 국가, 차량 모델, 연도, 월별로 그룹화하여 판매량 합계 및 건수 확인
grouped = df_all.groupby(['국가', '차량 모델', '연도', '월'])['판매량'].agg(['sum', 'count']).reset_index()

# 그룹별 0이 차지하는 비율(예를 들어, 모든 건수가 0인 경우)
grouped['zero_ratio'] = grouped.apply(lambda row: 1 if row['sum'] == 0 else 0, axis=1)

# 0인 그룹 확인 (sum이 0인 그룹)
zero_groups = grouped[grouped['sum'] == 0]
print(zero_groups)
```

```python
def replace_zero_with_min(group):
    # 0이 아닌 값들 추출
    nonzero = group[group > 0]
    # 만약 0이 아닌 값이 있으면 그 중 최솟값을 구하고, 그렇지 않으면 0 유지
    min_val = nonzero.min() if not nonzero.empty else 0
    # 0인 값을 min_val로 대체
    return group.replace(0, min_val)

# 국가와 차량 모델별로 판매량을 변환
df_all['판매량'] = df_all.groupby(['국가', '차량 모델'])['판매량'].transform(replace_zero_with_min)
```

```python
# 국가, 차량 모델, 연도, 월별로 그룹화하여 판매량 합계 및 건수 확인
grouped = df_all.groupby(['국가', '차량 모델', '연도', '월'])['판매량'].agg(['sum', 'count']).reset_index()

# 그룹별 0이 차지하는 비율(예를 들어, 모든 건수가 0인 경우)
grouped['zero_ratio'] = grouped.apply(lambda row: 1 if row['sum'] == 0 else 0, axis=1)

# 0인 그룹 확인 (sum이 0인 그룹)
zero_groups = grouped[grouped['sum'] == 0]
print(zero_groups)
```

```python
# 그룹별 판매량 합계가 0이 아닌 그룹만 남기기
df_all_nonzero = df_all.groupby(['국가', '차량 모델', '연도', '월']).filter(lambda group: group['판매량'].sum() != 0)

# 결과 확인
print(df_all_nonzero.groupby(['국가', '차량 모델', '연도', '월'])['판매량'].sum().reset_index())
```
