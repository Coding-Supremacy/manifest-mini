```python
def categorize_model(model):
    model_lower = model.lower()
    
    # 화물차: 이름에 'truck', 'bus', 'porter', 'mighty', 'tasman', 'bongo' 등이 포함되면
    if any(keyword in model_lower for keyword in ['truck', 'bus', 'porter', 'mighty', 'tasman', 'bongo']):
        return '화물차'
    
    # 경차: 주로 'santro', 'xcent', 'grand i10', 'i10', 'morning', 'picanto'
    if any(keyword in model_lower for keyword in ['santro', 'xcent', 'grand i10', 'i10', 'morning', 'picanto']):
        return '경차'
    
    # 세단: 'aura', 'solaris', 'mistra', 'la festa', 'reina', 'verna', 'elantra', 'sonata', 'accent', 'rio', 'optima', 'cadenza', 'celesta', 'stinger', 'k9'
    if any(keyword in model_lower for keyword in ['aura', 'solaris', 'mistra', 'la festa', 'reina', 'verna', 'elantra', 'sonata', 'accent', 'rio', 'optima', 'cadenza', 'celesta', 'stinger', 'k9']):
        return '세단'
    
    # SUV: 이름에 'suv' 또는 'cuv'가 있거나, SUV 관련 키워드가 포함된 경우
    if 'suv' in model_lower or 'cuv' in model_lower or any(keyword in model_lower for keyword in [
        'telluride','sorento','seltos','sportage','palisade','soul','k7 / k8','ev','tucson','santa fe','santa-fe','gv70','creta','kona','ix25','ix35','venue','ioniq5','sonet','stonic','niro','kx1','kx3','kx7'
    ]):
        return 'SUV'
    
    # IONIQ6는 전기 세단으로 분류
    if 'ioniq6' in model_lower:
        return '세단'
    
    # 추가 경차 조건: 'nios', 'i20', 'i10'
    if any(keyword in model_lower for keyword in ['nios', 'i20', 'i10']):
        return '경차'
    
    # 그 외는 기타로 분류
    return '기타'

# "차량 모델" 컬럼을 기반으로 "카테고리" 컬럼 생성 (매핑)
df_all['카테고리'] = df_all['차량 모델'].apply(categorize_model)

# 결과 확인
df_all
```

```python
import numpy as np

# 반복되는 판매량을 감지해서 ±3% 노이즈 추가하는 함수
def smooth_repeating_in_group(group):
    group = group.sort_values('날짜')
    sales = group['판매량'].tolist()
    new_sales = []
    prev = None
    for s in sales:
        if s == prev:
            noise = np.random.uniform(-0.50, 0.50) * s
            new_sales.append(int(s + noise))
        else:
            new_sales.append(s)
        prev = s
    group['판매량'] = new_sales
    return group

# df_total에 적용
df_total_smooth = df_total.groupby(['국가', '카테고리', '브랜드']).apply(smooth_repeating_in_group).reset_index(drop=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

def forecast_with_train_test_split(selected_brand, selected_category, selected_country):
    """
    1) df에서 브랜드, 카테고리, 국가를 필터링
    2) 날짜(ds)와 판매량(y)만 추출 및 정렬
    3) 마지막 3개월을 테스트 세트로 분할
    4) 학습 + 테스트 길이만큼 예측 -> 테스트 기간 예측과 실제값 비교하여 MAPE 계산
    5) 추가로 미래 12개월 예측 -> 시각화
    """

    grouped_df = df_total_smooth.groupby(
        ['국가', '연도', '월', '날짜', '브랜드', '카테고리'],
        as_index=False
    )['판매량'].sum()
    
    df_filtered = grouped_df[
        (grouped_df['브랜드'] == selected_brand) &
        (grouped_df['카테고리'] == selected_category) &
        (grouped_df['국가'] == selected_country)
    ]

    # 2. Prophet용 데이터 준비: ds, y
    df_model = df_filtered[['날짜', '판매량']].copy()
    df_model = df_model.rename(columns={'날짜': 'ds', '판매량': 'y'})
    df_model['ds'] = pd.to_datetime(df_model['ds'])
    df_model = df_model.sort_values('ds').reset_index(drop=True)

    # 데이터 최소 행 개수 확인
    if df_model.shape[0] < 2:
        raise ValueError("선택한 조건에 해당하는 데이터가 충분하지 않습니다.")

    # 3. 학습/테스트 분할 (마지막 3개월을 테스트 세트)
    split_date = df_model['ds'].max() - pd.DateOffset(months=3)
    train_df = df_model[df_model['ds'] <= split_date].copy()
    test_df = df_model[df_model['ds'] > split_date].copy()

    # 4. Prophet 모델 학습 (월별 데이터 → weekly/daily 시즈널리티 OFF)
    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False
    )
    model.fit(df_model)  # 전체 데이터로 학습

    # 5. 테스트 기간 길이만큼 미래 예측 -> 테스트 예측값 추출
    n_test = test_df.shape[0]
    future_test = model.make_future_dataframe(periods=n_test, freq='MS')
    forecast_test_all = model.predict(future_test)

    # 테스트 기간 예측값만 추출
    forecast_test = forecast_test_all[forecast_test_all['ds'] > split_date].copy()

    # 실제값, 예측값
    actual_test = test_df['y'].values
    predicted_test = forecast_test['yhat'].values

    # 6. 미래 12개월 예측
    future_12 = model.make_future_dataframe(periods=12, freq='MS')
    forecast_12 = model.predict(future_12)

    # 7. 테스트 기간 날짜 목록 기준으로 다시 예측값/실제값 매칭
    test_dates = test_df['ds'].unique()
    forecast_test_matched = forecast_12[forecast_12['ds'].isin(test_dates)].copy()
    forecast_test_matched = forecast_test_matched.sort_values('ds').reset_index(drop=True)
    test_df_sorted = test_df.sort_values('ds').reset_index(drop=True)

    if len(forecast_test_matched) == len(test_df_sorted):
        actual = test_df_sorted['y'].values
        predicted = forecast_test_matched['yhat'].values

        # 0 제외한 MAPE 계산
        nonzero_idx = actual != 0
        if nonzero_idx.sum() == 0:
            mape = np.nan
        else:
            mape = np.mean(np.abs((actual[nonzero_idx] - predicted[nonzero_idx]) / actual[nonzero_idx])) * 100
    else:
        # 날짜가 정확히 매칭되지 않는 경우, fallback
        actual = actual_test
        predicted = predicted_test
        nonzero_idx = actual != 0
        if nonzero_idx.sum() == 0:
            mape = np.nan
        else:
            mape = np.mean(np.abs((actual[nonzero_idx] - predicted[nonzero_idx]) / actual[nonzero_idx])) * 100

    print("Manual MAPE: {:.2f}%".format(mape))

    # 8. 예측 결과 시각화 (미래 12개월 예측 그래프)
#한글 폰트 설정
    plt.rc('font', family='NanumGothic')

    fig = model.plot(forecast_12)
    plt.rc('font', family='NanumGothic')
    plt.title(f"{selected_brand} / {selected_category} / {selected_country} 판매량 예측 (Train/Test split)")
    plt.xlabel("날짜")
    plt.ylabel("판매량")
    plt.show()

    return model, forecast_12

# 사용 예시
selected_brand = '기아'
selected_category = '세단'
selected_country = '북미-캐나다'

model, forecast = forecast_with_train_test_split(selected_brand, selected_category, selected_country)
```
