```python
def categorize_model(model):
    model_lower = model.lower()
    
    # 화물차: 이름에 'truck', 'bus', 'porter', 'mighty', 'tasman', 'bongo' 등이 포함되면
    if any(keyword in model_lower for keyword in ['truck', 'bus', 'porter', 'mighty', 'tasman', 'bongo']):
        return '화물차'
    
    # 경차: 주로 'santro', 'xcent', 'grand i10', 'i10', 'morning', 'picanto'
    if any(keyword in model_lower for keyword in ['santro', 'xcent', 'grand i10', 'i10', 'morning', 'picanto']):
        return '경차'
    
    # 세단: 'aura', 'solaris', 'mistra', 'la festa', 'reina', 'verna', 'elantra', 'sonata', 'accent', 'rio', 'optima', 'cadenza', 'celesta', 'stinger', 'k9'
    if any(keyword in model_lower for keyword in ['aura', 'solaris', 'mistra', 'la festa', 'reina', 'verna', 'elantra', 'sonata', 'accent', 'rio', 'optima', 'cadenza', 'celesta', 'stinger', 'k9']):
        return '세단'
    
    # SUV: 이름에 'suv' 또는 'cuv'가 있거나, SUV 관련 키워드가 포함된 경우
    if 'suv' in model_lower or 'cuv' in model_lower or any(keyword in model_lower for keyword in [
        'telluride','sorento','seltos','sportage','palisade','soul','k7 / k8','ev','tucson','santa fe','santa-fe','gv70','creta','kona','ix25','ix35','venue','ioniq5','sonet','stonic','niro','kx1','kx3','kx7'
    ]):
        return 'SUV'
    
    # IONIQ6는 전기 세단으로 분류
    if 'ioniq6' in model_lower:
        return '세단'
    
    # 추가 경차 조건: 'nios', 'i20', 'i10'
    if any(keyword in model_lower for keyword in ['nios', 'i20', 'i10']):
        return '경차'
    
    # 그 외는 기타로 분류
    return '기타'

# "차량 모델" 컬럼을 기반으로 "카테고리" 컬럼 생성 (매핑)
df_all['카테고리'] = df_all['차량 모델'].apply(categorize_model)

# 결과 확인
df_all
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics

def forecast_export(selected_brand, selected_category, selected_country):
    # 1. 데이터 필터링: 브랜드, 카테고리, 국가 기준으로 필터링
    df_filtered = df_total.copy()
    df_filtered = df_filtered[df_filtered['브랜드'] == selected_brand]
    df_filtered = df_filtered[df_filtered['카테고리'] == selected_category]
    df_filtered = df_filtered[df_filtered['국가'] == selected_country]
    
    # 2. 날짜별 판매량(수출량) 합산
    df_grouped = df_filtered.groupby('날짜', as_index=False).agg({'판매량': 'sum'})
    
    # 3. Prophet 요구사항에 맞게 컬럼명 변경: 날짜 -> ds, 판매량 -> y
    df_grouped = df_grouped.rename(columns={'날짜': 'ds', '판매량': 'y'})
    df_grouped['ds'] = pd.to_datetime(df_grouped['ds'])
    
    # 데이터가 충분한지 확인 (최소 2개 이상의 데이터가 필요)
    if df_grouped.shape[0] < 2:
        raise ValueError("선택한 조건에 해당하는 데이터가 충분하지 않습니다.")
    
    # 4. Prophet 모델 학습
    model = Prophet()
    model.fit(df_grouped)
    
    df_cv = cross_validation(model, initial='365 days', period='60 days', horizon='90 days')
    df_p = performance_metrics(df_cv)
    mape = df_p['mape'].mean()
    print("MAPE: {:.2f}%".format(mape * 100))
    
    # 5. 미래 예측: 365일 후까지 예측
    future = model.make_future_dataframe(periods=90)
    forecast = model.predict(future)
    
    return model, forecast

# 사용 예시
selected_brand = '현대'
selected_category = '세단'
selected_country = '북미-미국'

model, forecast = forecast_export(selected_brand, selected_category, selected_country)

#한글 폰트 설정
plt.rc('font', family='NanumGothic')
# 예측 결과 시각화
fig = model.plot(forecast)
plt.title(f"{selected_brand} / {selected_category} / {selected_country} 예측 결과")
plt.xlabel("날짜")
plt.ylabel("판매량(수출량)")
plt.show()ㅁ
```
