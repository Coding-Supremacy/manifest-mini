```python
import pandas as pd
```

```python
df=pd.read_csv('../data/전처리 완료/현대_차종별판매실적.csv')
```

```python
df
```

```python
df_domestic = df[df['판매 구분'] == '내수용'].copy()
```

```python
df_domestic
```

```python
df_foreign = df[df['판매 구분'] == '수출용'].copy()
```

```python
# 1월~12월 판매량 합산 → '총판매'
months = [f'{i}월' for i in range(1, 13)]
df_domestic['총판매'] = df_domestic[months].sum(axis=1)
df_foreign['총판매'] = df_foreign[months].sum(axis=1)
```

```python
df_domestic
```

```python
df_foreign
```

```python
domestic_sales = df_domestic.groupby('차량 모델')['총판매'].sum().reset_index()
domestic_sales.columns = ['차량 모델', '총판매_내수']
```

```python
domestic_sales
```

```python
foreign_sales = df_foreign.groupby('차량 모델')['총판매'].sum().reset_index()
foreign_sales.columns = ['차량 모델', '총판매_외수']
```

```python
df_model = pd.merge(domestic_sales, foreign_sales, on='차량 모델', how='outer').fillna(0)
```

```python
df_model[df_model['차량 모델']=='Santa-Fe (TMa)']
```

```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rc('font', family='NanumGothic')

# 내수 판매량 기준으로 DataFrame 정렬 (내림차순)
df_sorted = df_model.sort_values("총판매_내수", ascending=False).reset_index(drop=True)

# x축 순서를 재설정
x = range(len(df_sorted))

plt.figure(figsize=(12, 6))
plt.bar(x, df_sorted["총판매_내수"], width=0.4, label="내수 판매", align="edge")
plt.bar(x, df_sorted["총판매_외수"], width=-0.4, label="외수 판매", align="edge")
plt.xticks(x, df_sorted["차량 모델"], rotation=90, fontsize=8)
plt.title("23년~25년 모델별 내수 vs 해외 판매 비교 (내수 판매 기준 정렬)")
plt.xlabel("차량 모델")
plt.ylabel("총 판매량")
plt.legend()
plt.tight_layout()
plt.savefig('../image/모델별 내수 vs 해외 판매 비교(내수 판매 기준).png')
plt.show()
```

```python
df=pd.read_csv('../data/전처리 완료/현대_차종별판매실적.csv')
```

```python
df_domestic = df[df['판매 구분'] == '내수용'].copy()
```

```python
df_foreign = df[df['판매 구분'] == '수출용'].copy()
```

```python
# 1월~12월 판매량 합산 → '총판매'
months = [f'{i}월' for i in range(1, 13)]
df_domestic['총판매'] = df_domestic[months].sum(axis=1)
df_foreign['총판매'] = df_foreign[months].sum(axis=1)
```

```python
domestic_yearly_sales = df_domestic.groupby('연도')['총판매'].sum().reset_index()
foreign_yearly_sales = df_foreign.groupby('연도')['총판매'].sum().reset_index()
```

```python
df_yearly = pd.merge(domestic_yearly_sales, foreign_yearly_sales, on='연도', how='outer', suffixes=('_내수', '_외수'))
df_yearly = df_yearly.sort_values(by='연도')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_yearly['연도'], df_yearly['총판매_내수'], marker='o', label='내수 판매')
plt.plot(df_yearly['연도'], df_yearly['총판매_외수'], marker='s', label='외수 판매')
plt.title("연도별 내수 vs 외수 판매량 추이")
plt.xlabel("연도")
plt.ylabel("총 판매량")
plt.legend()
plt.grid(True)
plt.show()
```
