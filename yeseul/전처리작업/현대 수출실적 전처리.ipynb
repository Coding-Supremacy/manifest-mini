```python
df25=pd.read_csv('../data/현대/hmc-export-by-region-january-y2025.csv')
```

```python
month_mapping = {
    "Jan.": "1월",
    "Feb.": "2월",
    "Mar.": "3월",
    "Apr.": "4월",
    "May": "5월",
    "Jun.": "6월",
    "Jul.": "7월",
    "Aug.": "8월",
    "Sep.": "9월",
    "Oct.": "10월",
    "Nov.": "11월",
    "Dec.": "12월"
}
df25.rename(columns=month_mapping, inplace=True)


df25["연도"] = 2025
df25.rename(columns={"Region":"국가","Continent":"대륙"}, inplace=True)
```

```python
df_all=pd.concat([df23,df24,df25])
```

```python
df_all.dtypes
```

```python
import re
import numpy as np

month_cols = [f"{i}월" for i in range(1, 13)]

# 괄호 숫자 (예: '(1)')를 0으로 바꾸고, 나머지는 float으로 변환
def clean_value(val):
    val = str(val).strip()
    if re.match(r'^\(\d+\)$', val):  # 괄호 숫자 → 0
        return 0
    if val in ["-", "None", "nan", ""]:
        return np.nan
    val = val.replace(",", "").replace("–", "").replace("−", "").replace("·", "").replace("\xa0", "")
    try:
        return float(val)
    except ValueError:
        return np.nan

# 월별 컬럼에 적용
for col in month_cols:
    df_all[col] = df_all[col].apply(clean_value)

# 최종 데이터 타입 확인
df_all.dtypes
```

```python
df_all.fillna(0, inplace=True)
```

```python
df_all['대륙'] = df_all['대륙'].str.replace('\n', ' ', regex=False)
```
