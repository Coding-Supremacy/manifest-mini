```python
df_hyundai_sales=pd.concat([df23eu, df24eu, df25eu, df23us, df24us, df25us])
```

```python
df_hyundai_sales.dtypes
```

```python
import re
import numpy as np

month_cols = [f"{i}월" for i in range(1, 13)]

# 괄호 숫자 (예: '(1)')를 0으로 바꾸고, 나머지는 float으로 변환
def clean_value(val):
    val = str(val).strip()
    if re.match(r'^\(\d+\)$', val):  # 괄호 숫자 → 0
        return 0
    if val in ["-", "None", "nan", ""]:
        return np.nan
    val = val.replace(",", "").replace("–", "").replace("−", "").replace("·", "").replace("\xa0", "")
    try:
        return float(val)
    except ValueError:
        return np.nan

# 월별 컬럼에 적용
for col in month_cols:
    df_hyundai_sales[col] = df_hyundai_sales[col].apply(clean_value)

# 최종 데이터 타입 확인
df_hyundai_sales.dtypes
```

```python
df_hyundai_sales.head()
```

```python
df_hyundai_sales.fillna(0, inplace=True)
```

```python
df_all=df_hyundai_sales
```

```python
type_map = {
    'PC': '승용차',
    'RV': '레저용 차량',
    'Light CV': '경상용차'
}

df_all['차량 구분'] = df_all['차량 구분'].map(type_map)
```

```python
df_all.rename(columns={'Models':"차량 모델"},inplace=True)
```
