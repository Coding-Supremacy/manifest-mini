```python
import pandas as pd
```

```python
df=pd.read_csv('../../data/기아_지역별수출실적_전처리.csv')
```

```python
df['국가명'].unique()
```

```python
df=df.groupby(['국가명','연도'])[['1월', '2월', '3월', '4월', '5월', '6월','7월', '8월', '9월', '10월', '11월', '12월']].sum().reset_index()
```

```python
df
```

```python
df2=pd.read_csv('기아_21_22년_지역별.csv')
```

```python
df2
```

```python
month_cols = ['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월','연도']
df2[month_cols] = df2[month_cols].replace(',', '', regex=True).astype(int)
```

```python
df2
```

```python
df_combined = pd.concat([df, df2], ignore_index=True)
```

```python
df_combined
```

```python
df_melted=pd.melt(df_combined, id_vars=['국가명','연도'], value_vars=['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월'], var_name='월', value_name='수출실적')
```

```python
df_melted
```

```python
df_melted['월']=df_melted['월'].str.replace('월','').astype(int)
```

```python
df_melted
```

```python
df_melted['ds']=pd.to_datetime(df_melted['연도'].astype(str)+'-'+df_melted['월'].astype(str).str.zfill(2)+'-01')
```

```python
df_melted.rename(columns={'수출실적':'y'}, inplace=True)
```

```python
df_melted
```

```python
#ds가 2025-02-01 이후인 데이터는 제거
df_melted=df_melted[df_melted['ds']<'2025-02-01']
```

```python
#날짜 정렬
df_melted=df_melted.sort_values('ds')
```

```python
df_melted
```

```python
df_melted[df_melted['국가명']=='북미-미국']
```

```python
df_melted.to_csv('기아_시장구분별_수출실적.csv', index=False)
```

```python
import os
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import pickle

# 그래프 및 모델 저장 디렉토리 설정
os.makedirs("prophet_forecasts", exist_ok=True)
os.makedirs("prophet_models", exist_ok=True)

# 국가명 리스트 추출
countries = df_melted["국가명"].unique()

# 국가별로 Prophet 모델 적용
for country in countries:
    df_country = df_melted[df_melted["국가명"] == country][["ds", "y"]].copy()
    
    # Prophet 모델 생성 및 학습
    model = Prophet(daily_seasonality=False, weekly_seasonality=False)
    model.fit(df_country)

    # 향후 18개월 예측
    future = model.make_future_dataframe(periods=18, freq="MS")
    forecast = model.predict(future)

    # 📈 예측 그래프 저장
    fig = model.plot(forecast)
    plt.title(f"[{country}] 수출실적 예측")
    plt.savefig(f"prophet_forecasts/kia_{country}_forecast.png")
    plt.show()
    plt.close(fig)  # 메모리 누수 방지
    

    # 💾 모델 저장 (Pickle 형식)
    with open(f"prophet_models/kia_{country}_model.pkl", "wb") as f:
        pickle.dump(model, f)

    print(f"[완료] {country} - 그래프 및 모델 저장")
```

```python
last_date = df_country['ds'].max()
future_forecast = forecast[forecast['ds'] > last_date]

future_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
future_forecast.to_csv('기아_북미미국_수출실적_미래예측.csv', index=False)
```
