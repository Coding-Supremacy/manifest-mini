```python
# 변환하려는 컬럼 목록
cols_to_convert = ['1월', '2월', '3월', '4월', '5월', '6월',
                   '7월', '8월', '9월', '10월', '11월', '12월']

for col in cols_to_convert:
    # 각 셀을 문자열로 변환한 후 쉼표 제거
    df_factory[col] = df_factory[col].apply(lambda x: str(x).replace(',', ''))
    # 숫자로 변환 (변환 불가능한 경우 NaN 처리)
    df_factory[col] = pd.to_numeric(df_factory[col], errors='coerce')
    # NaN을 0으로 채우고 정수형으로 변환
    df_factory[col] = df_factory[col].fillna(0).astype(int)

df_factory.dtypes
```

```python
# 변환하려는 컬럼 목록
cols_to_convert = ['1월', '2월', '3월', '4월', '5월', '6월',
                   '7월', '8월', '9월', '10월', '11월', '12월']

for col in cols_to_convert:
    # 각 셀을 문자열로 변환한 후 쉼표 제거
    df_sales[col] = df_sales[col].apply(lambda x: str(x).replace(',', ''))
    # 숫자로 변환 (변환 불가능한 경우 NaN 처리)
    df_sales[col] = pd.to_numeric(df_sales[col], errors='coerce')
    # NaN을 0으로 채우고 정수형으로 변환
    df_sales[col] = df_sales[col].fillna(0).astype(int)

df_sales.dtypes
```

```python
months = [f'{i}월' for i in range(1, 13)]
df_factory['총생산량'] = df_factory[months].sum(axis=1)
# 모델별 총 생산량 집계 (공장별로 여러 행이 있을 수 있으므로)
factory_production = df_factory.groupby('차량 모델')['총생산량'].sum().reset_index()
```

```python
factory_production
```

```python
df_sales['총판매량'] = df_sales[months].sum(axis=1)
```

```python
sales_by_model = df_sales.groupby('차량 모델')['총판매량'].sum().reset_index()
```

```python

name_mapping = {
    "bongo": "bongo",
    "bus": "bus",
    "carnival": "carnival",
    "carnival / sedona": "carnival",
    "ev3": "ev3",
    "ev4": "ev4",
    "ev5": "ev5",
    "ev6": "ev6",
    "ev9": "ev9",
    "k3": "k3",
    "k5": "k5",
    "k5 / optima": "k5",
    "k7 / k8": "k8",
    "k7 / k8 / cadenza": "k8",
    "k9": "k9",
    "military": "military",
    "millitary": "military",  # 오타 보정
    "mohave": "mohave",
    "mohave / borrego": "mohave",
    "morning": "morning",
    "morning / picanto": "morning",
    "new carens": "new carens",
    "niro": "niro",
    "niro plus": "niro plus",
    "pv5": "pv5",
    "pride / rio": "pride",
    "ray": "ray",
    "seltos": "seltos",
    "sorento": "sorento",
    "soul": "soul",
    "sportage": "sportage",
    "stinger": "stinger",
    "stonic": "stonic",
    "tasman": "tasman",
    "accent (hc)": "accent",
    "c'eed": "ceed",
    "c'eed cuv": "ceed",
    "carens mpv": "carens",
    "carnival (ckd)": "carnival",
    "cerato": "cerato",
    "forte": "forte",
    "k2": "k2",
    "k4": "k4",
    "kx1": "kx1",
    "kx3": "kx3",
    "kx5": "kx5",
    "kx7": "kx7",
    "optima / k5": "k5",
    "pegas": "pegas",
    "rio(pride)": "pride",
    "sonet": "sonet",
    "syros": "syros",
    "telluride": "telluride",
    "tucson(nx4)": "tucson",
    "zhipao": "zhipao"
}

def normalize_model_name(name):

    # 이름의 앞뒤 공백 제거, 소문자 변환 후 사전(mapping) 적용
    norm = str(name).strip().lower()
    # 공백이 여러 개 있는 경우 단일 공백으로 치환
    norm = ' '.join(norm.split())
    return name_mapping.get(norm, norm)

def standardize_model_names(df, col_name='차량 모델'):
    # 각 셀에 normalize_model_name 적용
    df[col_name] = df[col_name].apply(normalize_model_name)
    return df

factory_production = standardize_model_names(factory_production, col_name='차량 모델')
sales_by_model = standardize_model_names(sales_by_model, col_name='차량 모델')

# 결과 확인
print(factory_production['차량 모델'].unique())
print(sales_by_model['차량 모델'].unique())
```
