```python
# 영어 → 한글 매핑 딕셔너리
country_mapping = {
    'US': '북미-미국',
    'Canada': '북미-캐나다',
    'Mexico': '북미-멕시코',
    'EU+EFTA': '서유럽',
    'E.Europe/CIS': '동유럽',
    'Latin America': '중남미',
    'Middle East/Africa': '중동·아프리카',
    'Asia / Pacific': '아시아',
    'China': '아시아',
    'India': '기타'
}

# 매핑 적용
df['국가명'] = df['국가명'].map(country_mapping)
```
