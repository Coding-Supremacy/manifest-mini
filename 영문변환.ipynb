```python
# 영어 → 한글 매핑 딕셔너리
country_mapping = {
    'US': '북미-미국',
    'Canada': '북미-캐나다',
    'Mexico': '북미-멕시코',
    'EU+EFTA': '서유럽',
    'E.Europe/CIS': '동유럽',
    'Latin America': '중남미',
    'Middle East/Africa': '중동·아프리카',
    'Asia / Pacific': '아시아',
    'China': '아시아',
    'India': '기타'
}

# 매핑 적용
df['국가명'] = df['국가명'].map(country_mapping)
```

```python
df['국가명']
```

```python
df[df['국가명']=='아시아']
```

```python
df
```

```python
# 📌 월 컬럼 리스트
months = [f'{i}월' for i in range(1, 13)]

# 🎯 국가명 기준으로 월별 데이터 합산
df_grouped = df.groupby(['차량 구분', '국가명', '연도'])[months].sum().reset_index()
```

```python
df_grouped
```

```python
df_grouped['국가명'].unique()
```

```python
df_grouped
```

```python
df_grouped.to_csv('data/기아_지역별수출실적_전처리.csv', index=False)
```
