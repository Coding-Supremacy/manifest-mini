```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# seaborn 스타일 설정
sns.set(style="whitegrid")
```

```python
# 1. 공장별 전체 판매량 추이
plt.figure(figsize=(12, 5))
df_group1 = df_melted.groupby(["날짜", "공장명(국가)"])["판매량"].sum().reset_index()
sns.lineplot(data=df_group1, x="날짜", y="판매량", hue="공장명(국가)", marker="o")
plt.title("공장별 전체 판매량 추이")
plt.xlabel("날짜")
plt.ylabel("판매량")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2. 차량 모델별 판매량 추이 (상위 3개 모델)
top_models = df_melted["차량 모델"].value_counts().head(3).index.tolist()
df_top_models = df_melted[df_melted["차량 모델"].isin(top_models)]
```

```python
plt.figure(figsize=(12, 5))
df_group2 = df_top_models.groupby(["날짜", "차량 모델"])["판매량"].sum().reset_index()
sns.lineplot(data=df_group2, x="날짜", y="판매량", hue="차량 모델", marker="o")
plt.title("차량 모델별 판매량 추이 (상위 3개 모델)")
plt.xlabel("날짜")
plt.ylabel("판매량")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 3. 내수용 vs 수출용 비교
plt.figure(figsize=(12, 5))
df_group3 = df_melted.groupby(["날짜", "판매 구분"])["판매량"].sum().reset_index()
sns.lineplot(data=df_group3, x="날짜", y="판매량", hue="판매 구분", marker="o")
plt.title("내수용 vs 수출용 판매량 추이")
plt.xlabel("날짜")
plt.ylabel("판매량")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 4. 공장 + 모델 조합별 Top 3 분석
df_group4 = df_melted.groupby(["공장명(국가)", "차량 모델"])["판매량"].sum().reset_index()
top_combinations = df_group4.sort_values(by="판매량", ascending=False).head(3)[["공장명(국가)", "차량 모델"]]
```

```python
# 해당 조합만 필터링
df_top_combinations = df_melted.merge(top_combinations, on=["공장명(국가)", "차량 모델"])
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=df_top_combinations, x="날짜", y="판매량", hue="차량 모델", style="공장명(국가)", marker="o")
plt.title("공장 + 모델 조합별 판매 추이 (Top 3)")
plt.xlabel("날짜")
plt.ylabel("판매량")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def analyze_factory_sales(file_path):
    df = pd.read_csv(file_path)
    df_melted = df.melt(id_vars=["공장명(국가)", "차량 모델", "판매 구분", "연도"],
                        value_vars=[f"{i}월" for i in range(1, 13)],
                        var_name="월", value_name="판매량")
    df_melted["월"] = df_melted["월"].str.replace("월", "").astype(int)
    df_melted["날짜"] = pd.to_datetime(df_melted["연도"].astype(str) + "-" + df_melted["월"].astype(str) + "-01")
    df_grouped = df_melted.groupby("날짜")["판매량"].sum().reset_index()
    df_grouped["증가율"] = df_grouped["판매량"].pct_change() * 100
    return df_grouped
```

```python
# ---------- 마케팅 타이밍 추천 함수 ----------
def recommend_marketing_timing(df_grouped, column_name="판매량", threshold=20):
    df_grouped = df_grouped.copy()
    df_grouped["전월"] = df_grouped[column_name].shift(1)
    df_grouped["증가율"] = (df_grouped[column_name] - df_grouped["전월"]) / df_grouped["전월"] * 100
    recommendations = df_grouped[df_grouped["증가율"] >= threshold][["날짜", column_name, "증가율"]]
    return recommendations.sort_values("날짜")

# ---------- 1. 공장별 판매실적 분석 ----------
def analyze_factory_sales(file_path):
    df = pd.read_csv(file_path)
    df_melted = df.melt(id_vars=["공장명(국가)", "차량 모델", "판매 구분", "연도"],
                        value_vars=[f"{i}월" for i in range(1, 13)],
                        var_name="월", value_name="판매량")
    df_melted["월"] = df_melted["월"].str.replace("월", "").astype(int)
    df_melted["날짜"] = pd.to_datetime(df_melted["연도"].astype(str) + "-" + df_melted["월"].astype(str) + "-01")
    df_grouped = df_melted.groupby("날짜")["판매량"].sum().reset_index()
    df_grouped["증가율"] = df_grouped["판매량"].pct_change() * 100
    return df_grouped, recommend_marketing_timing(df_grouped, "판매량")
```
